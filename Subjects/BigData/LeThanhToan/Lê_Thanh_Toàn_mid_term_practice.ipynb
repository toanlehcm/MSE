### II. Using the same data from assignment I, do the following requirements:

**1. Split the DataFrame into `X` (the data) and `y` (the labels).** </br>
**2. Using `train_test_split`, split `X` and `y` into training and test sets `(X_train, X_test, y_train, and y_test)`.
Set the random number generator state to 0 using `random_state=0` to make sure your results match the autograder!**

```python
# II.1
X = cancer_df.iloc[:,0:30].values
Y = cancer_df.iloc[:,-1].values
X.shape
```

```python
# Or
```

```python
# II.2
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state = 0) 
```

### III. Using pipeline in `sklearn` to stack the data pipeline including `StandardScaler`, `PCA` (with `n_components=7`) and using `LogisticRegression` to build the ML model. Then compute and print out the test accuracy

```python
# III
```

```python
from sklearn.preprocessing import  StandardScaler
from sklearn.decomposition import  PCA
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import  make_pipeline
from sklearn.metrics import accuracy_score
```

```python
data_pipeline = make_pipeline(StandardScaler(), PCA(n_components=7), LogisticRegression())
data_pipeline.fit(X_train, y_train)
y_pred = data_pipeline.predict(X_test)
acc = accuracy_score(y_test, y_pred)

print('Accuracy of out model is queal: ' + str(round(acc, 2)) + '%.')
```

```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_digits

dataset = load_digits()
X, y = dataset.data, dataset.target
print(X.shape)
```

```python
import matplotlib.pyplot as plt 
_, axes = plt.subplots(nrows=1, ncols=8, figsize=(15, 8))
for ax, image, label in zip(axes, dataset.images, dataset.target):
    ax.set_axis_off()
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')
    ax.set_title('Training: %i' % label)
```

```python
# V.1
dataset.keys()
digits_class = pd.DataFrame(X)
digits_class['class'] = pd.Series(dataset.target)
digits_class.groupby(['class']).size()
```

```python
# V.2
digits_class['binary_class'] = digits_class.apply(lambda x: 1 if x['class'] == 1 else 0, axis=1)
digits_class.groupby(['binary_class']).size()
```

```python
# V.3
from sklearn.svm import  SVC
from imblearn.metrics import  classification_report_imbalanced
```

```python
y_binary = digits_class.iloc[:,-1]

X_train, X_test, y_train, y_test = train_test_split(X, y_binary, test_size = 0.2, random_state = 0)
pipeline_SVM = make_pipeline(SVC())
pipeline_SVM.fit(X_train, y_train)

y_pred = pipeline_SVM.predict(X_test)

print(classification_report_imbalanced(y_test, y_pred))
```
